```python
#dataset inlezen
df = pd.read_csv('Students.csv')
```

Bepaal de verdeling tussen het aantal mannen en vrouwen in deze dataset. Maak vervolgens een staafdiagram die het aantal mannen en vrouwen weergeeft.

```python
mannen = df[df['sex'] == "M"].sex.count()
vrouwen = df[df['sex'] == "F"].sex.count()
bereik = abs(mannen - vrouwen)
klassenbreedte = round(bereik/np.sqrt(df.shape[0]),0)
bins = np.arange(start=vrouwen, stop=mannen + klassenbreedte , step=klassenbreedte)
sns.histplot(data=df,x="sex",bins=bins)
plt.title('histogram van geslacht')
```



### **Wat Gebeurt Hier?**

1. **Aantal Mannen en Vrouwen Bepalen:**

   * `mannen`: Telt alle rijen waar 'sex' gelijk is aan "M".
   * `vrouwen`: Telt alle rijen waar 'sex' gelijk is aan "F".

2. **Bereik Berekenen:**

   * **Bereik** is het verschil tussen het aantal mannen en vrouwen.

3. **Klassenbreedte Berekenen:**

   * De breedte van elke klasse (bin) wordt berekend met de formule:

   $$
   \text{Klassenbreedte} = \frac{\text{Bereik}}{\sqrt{n}}
   $$

   * Dit zorgt voor een evenwichtige verdeling van de bins.

4. **Bepalen van de Bin Grenswaarden:**

   * `np.arange()` genereert de grenzen voor de bins, van het aantal vrouwen tot het aantal mannen, met stappen van de berekende klassenbreedte.

5. **Histogram Plotten:**

   * `sns.histplot()` wordt gebruikt om de geslachtsverdeling te visualiseren.

6. **Grafiektitel Toevoegen:**

   * `plt.title()` voegt een titel toe om de grafiek te beschrijven.

Plot de histogram van de leeftijd. Bereken hiervoor zelf de klassebreedtes

```python
#howestmethode
maximum = df['age'].max()
minimum = df['age'].min()
klassenbreedte = np.ceil((maximum - minimum)/np.sqrt(df.shape[0]))
bins = np.arange(start=minimum, stop=maximum + klassenbreedte , step=klassenbreedte)
sns.histplot(data=df,x="age",bins=bins)
plt.title('histogram van leeftijd')
```

### **Uitleg van de Howestmethode voor Histogram van Leeftijd**

Deze code maakt een histogram van leeftijden in de dataset met de Howestmethode:

1. **Bepaal de hoogste en laagste leeftijd:**

   * Het **maximum** is de oudste persoon in de dataset.
   * Het **minimum** is de jongste persoon in de dataset.

2. **Bereken de klassenbreedte:**

   * Dit is de breedte van elke staaf (bin) in het histogram.
   * De formule hiervoor is:

   $$
   \text{Klassenbreedte} = \frac{\text{Maximum - Minimum}}{\sqrt{n}}
   $$

   * Hierbij is **n** het aantal personen in de dataset.
   * De functie **np.ceil()** rondt dit getal naar boven af, zodat alle leeftijden binnen de grenzen vallen.

3. **Maak de bins aan:**

   * Dit zijn de intervallen waarin de leeftijden gegroepeerd worden.
   * De bins beginnen bij de laagste leeftijd en gaan door tot de hoogste leeftijd plus één klassenbreedte.

4. **Plot het histogram:**

   * Gebruik **sns.histplot()** om de leeftijdsverdeling te visualiseren.
   * Voeg een titel toe met **plt.title()** om duidelijk te maken wat de grafiek voorstelt.

### **Uitleg per Lijn voor de Howestmethode (Final Grade Histogram)**

1. **Bepaal de Hoogste Waarde (Maximum):**

   * De eerste stap is om de hoogste eindscore (**G3**) in de dataset te vinden.
   * Dit is nodig om de volledige spreiding van de data te bepalen, zodat alle mogelijke scores in de grafiek worden opgenomen.

2. **Bepaal de Laagste Waarde (Minimum):**

   * Vervolgens wordt de laagste eindscore in de dataset bepaald.
   * Dit geeft het startpunt van de horizontale as in de histogram.

3. **Bereken de Klassenbreedte:**

   * De klassenbreedte bepaalt hoe breed elke staaf (bin) in de histogram is.
   * De formule die hier gebruikt wordt, verdeelt het verschil tussen de hoogste en laagste score door de wortel van het aantal rijen in de dataset.
   * Dit zorgt voor een redelijk aantal klassen, ongeacht de grootte van de dataset.
   * De functie **np.ceil()** wordt gebruikt om het resultaat naar boven af te ronden, zodat de klassenbreedte altijd een heel getal is.

4. **Maak de Bins aan:**

   * Hier worden de exacte grenzen voor elke klasse bepaald.
   * De bins beginnen bij de laagste waarde en gaan in stappen van de berekende klassenbreedte tot voorbij de hoogste waarde.
   * Dit zorgt ervoor dat elke mogelijke score in de dataset wordt opgenomen.

5. **Plot het Histogram:**

   * Ten slotte wordt de histogram zelf getekend met de eindcijfers als gegevensbron.
   * De titel "histogram van final grade" wordt toegevoegd om duidelijk te maken wat de grafiek voorstelt.


```python
import numpy as np
leeftijd = df['G3']
leeftijd_max = df['G3'].max()
leeftijd_min = df['G3'].min()
leeftijd_aantal = df['G3'].count()
resultaat_leeftijd = leeftijd_max - leeftijd_min
resultaat_leeftijd_aantal = leeftijd_aantal
klassenbreedte = resultaat_leeftijd/np.sqrt(resultaat_leeftijd_aantal)
bin_count = int((leeftijd_max - leeftijd_min) / klassenbreedte)
print(bin_count)

print(resultaat_leeftijd)
print(resultaat_leeftijd_aantal)
print(klassenbreedte)

plt.hist(df['G3'],bins=bin_count,edgecolor='blue')
plt.title('Histogram finale score')
plt.xlabel('G3 scores')
plt.ylabel('Aantal G3 scores')
plt.show()
```
