Plot de histogram van de leeftijd. Bereken hiervoor zelf de klassebreedtes

```python
import numpy as np
leeftijd = df['age']
leeftijd_max = df['age'].max()
leeftijd_min = df['age'].min()
leeftijd_aantal = df['age'].count()
resultaat_leeftijd = leeftijd_max - leeftijd_min
resultaat_leeftijd_aantal = leeftijd_aantal
klassenbreedte = resultaat_leeftijd/np.sqrt(resultaat_leeftijd_aantal)
bin_count = int((leeftijd_max - leeftijd_min) / klassenbreedte)
print(bin_count)

print(resultaat_leeftijd)
print(resultaat_leeftijd_aantal)
print(klassenbreedte)

plt.hist(df['age'],bins=bin_count,edgecolor='blue')
plt.title('Histogram leeftijd')
plt.xlabel('Leeftijd')
plt.ylabel('Aantal')
plt.show()

```

Bepaal telkens de centrummaten en de histogram van de *final grade*(G3) van mannen enerzijds en deze van vrouwen anderzijds.

```python
import numpy as np
import seaborn as sns
conditie_geslacht_mannen = df['sex'] == 'M'

max_mannen = df[conditie_geslacht_mannen].G3.max()
min_mannen = df[conditie_geslacht_mannen].G3.min()
aantal_mannen = df[conditie_geslacht_mannen].G3.count()
median_mannen = df[conditie_geslacht_mannen].G3.median()
mode_mannen = df[conditie_geslacht_mannen].G3.mode()
scheefheid_mannen = df[conditie_geslacht_mannen].G3.skew()
range_mannen = max_mannen - min_mannen
klassenbreedte = range_mannen/np.sqrt(aantal_mannen)
bin_count = int((max_mannen - min_mannen) / klassenbreedte)
print(bin_count)


plt.hist(df['G3'],bins=bin_count,edgecolor='blue')
plt.title('Histogram finale score ')
plt.xlabel('G3 scores mannen')
plt.ylabel('Aantal G3 scores')
plt.show()
```

```python
import numpy as np
import seaborn as sns
conditie_geslacht_vrouwen = df['sex'] == 'F'

max_vrouwen = df[conditie_geslacht_vrouwen].G3.max()
min_vrouwen = df[conditie_geslacht_vrouwen].G3.min()
aantal_vrouwen = df[conditie_geslacht_vrouwen].G3.count()
median_vrouwen = df[conditie_geslacht_vrouwen].G3.median()
mode_vrouwen = df[conditie_geslacht_vrouwen].G3.mode()
scheefheid_vrouwen = df[conditie_geslacht_vrouwen].G3.skew()
range_vrouwen = max_vrouwen - min_vrouwen
klassenbreedte = range_vrouwen/np.sqrt(aantal_vrouwen)
bin_count = int((max_vrouwen - min_vrouwen) / klassenbreedte)
print(bin_count)


plt.hist(df['G3'],bins=bin_count,edgecolor='blue')
plt.title('Histogram finale score ')
plt.xlabel('G3 scores vrouwen')
plt.ylabel('Aantal G3 scores')
plt.show()
```
