## <span style="color:deeppink"> Centrummaten en spreidingsmaten van de leeftijd

```python
gemiddelde = dataset['age'].mean()
print('gemiddelde = ', gemiddelde)
mediaan = dataset['age'].median()
print('mediaan = ', mediaan)
variantie =dataset['age'].var()
print('variantie = ', variantie)
standaardafwijking= dataset['age'].var()
print('standaardafwijking', standaardafwijking)
Q1= dataset['age'].quantile(0.25)
print('Q1 = ', Q1)
Q3= dataset['age'].quantile(0.75)
print('Q3 = ', Q3)
interkwartielafstand = Q3-Q1
print('Interkwartielafstand = ', interkwartielafstand)
variatiebreedte = dataset['age'].max()-dataset['age'].min()
print('Range = ',variatiebreedte)
scheefheid = dataset['age'].skew()
print('Scheefheid = ',scheefheid)
kurtosis = dataset['age'].kurt()
print('Kurtosis = ', kurtosis)
```

## <span style="color:deeppink"> Histogram van de leeftijd
Vervolgens zullen we een histogram maken met behulp van matplotlib om de verdeling van de leeftijd van de patienten weer te geven. Meer informatie over het plotten van een histogram en welke parameters aanwezig zijn kan je vinden via de volgende link: <br/>
https://seaborn.pydata.org/generated/seaborn.displot.html

```python
# Histogram van leeftijd
klassenbreedte = variatiebreedte/np.sqrt(dataset.age.count())
minimum = dataset.age.min()
maximum = dataset.age.max()
bins = np.arange(start = minimum, stop = maximum+round(klassenbreedte), step=round(klassenbreedte))
sns.displot(x=dataset['age'], bins =bins,kde = False)
plt.title("Histogram van de leeftijd")
```

## <span style="color:deeppink"> boxplot van de leeftijd </span>
