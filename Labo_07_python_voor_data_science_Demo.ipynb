```python
# zelf een dataframe maken uit een dictionary
mijn_team = {"Stijn": ["Frankrijk", "Zwitserland", "Oostenrijk", "Nederland"],
             "Christophe": ["USA", "Frankrijk", "Duitsland", "Nederland" ],
             "Dieter": ["Nederland", "Duitsland", "Zwitserland", "Oostenrijk"],
             "Gilles": ["UK", "Spanje", "Portugal", "Frankrijk"]}
```

Welke waarden zijn aanwezig in de kolom chestpain?  
<span style="color:DeepPink">  **Link naar de documentatie:** https://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.DataFrame.value_counts.html

```python
dataset['cp'].value_counts()
```

Voeg een kolom toe chol_te_hoog, wanneer de cholesterol waarde hoger is dan 300 dan wordt de waarde True ingevuld, anders de waarde False.  
<span style="color:DeepPink">  **Link naar hulp:** https://www.geeksforgeeks.org/ways-to-filter-pandas-dataframe-by-column-values/

```python
dataset['chol_te_hoog'] = dataset['chol'] > 300
dataset
```

Visualiseer het aantal personen met een te hoge cholesterol vs het aantal zonder te hoge cholesterol:  
<span style="color:DeepPink">  **Link naar de documentatie:** : https://seaborn.pydata.org/api.html

```python
sns.countplot(x='chol_te_hoog',data=dataset)
plt.title(" Cholesterol te hoog vs normaal")

plt.show()

```

Verwijder de kolom slope uit de dataset  
<span style="color:DeepPink">  **Link naar de documentatie:** https://pandas.pydata.org/docs/reference/api/pandas.DataFrame.drop.html

```python
dataset = dataset.drop('slope',axis=1)
```

Hoeveel personen met type 2 chestpain, hebben een te hoge cholesterol?  
<span style="color:DeepPink">  **Link naar hulp:**  https://www.geeksforgeeks.org/filter-pandas-dataframe-with-multiple-conditions/

```python
voorwaarde_1 = dataset['cp'] == 2
voorwaarde_2 = dataset['chol_te_hoog'] == True
dataset[voorwaarde_1&voorwaarde_2].cp.count()
```

# Gebruik maken van subplots
Maak een plot die de verschillen duidelijk weergeeft tussen de leeftijd van personen met chestpain type 0 en 1. Maak hierbij gebruik van subplots.
<span style="color:DeepPink">  **Link naar de documentatie:** :
- https://seaborn.pydata.org/api.html
- https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.subplots.html

```python
f, axes = plt.subplots(1,2)
sns.boxplot(y='age',data=dataset[dataset['cp'] == 0],whis = 1.5, ax = axes[0])
axes[0].set_title('Leeftijd chest type 0')
axes[0].set(ylim=(20,90))

sns.boxplot(y='age',data=dataset[dataset['cp'] == 1],whis = 1.5, ax = axes[1])
axes[1].set_title('Leeftijd chest type 1')
axes[1].set(ylim=(20,90))


```
